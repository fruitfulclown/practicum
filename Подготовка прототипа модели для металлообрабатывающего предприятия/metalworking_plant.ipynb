### Описание данных

**Технологический процесс**

- Rougher feed — исходное сырье
- Rougher additions (или reagent additions) — флотационные реагенты: Xanthate, Sulphate, Depressant
 - Xanthate **— ксантогенат (промотер, или активатор флотации);
 - Sulphate — сульфат (на данном производстве сульфид натрия);
 - Depressant — депрессант (силикат натрия).
- Rougher process (англ. «грубый процесс») — флотация
- Rougher tails — отвальные хвосты
- Float banks — флотационная установка
- Cleaner process — очистка
- Rougher Au — черновой концентрат золота
- Final Au — финальный концентрат золота

**Параметры этапов**

- air amount — объём воздуха
- fluid levels — уровень жидкости
- feed size — размер гранул сырья
- feed rate — скорость подачи

**Наименование признаков**
Наименование признаков должно быть такое:
[этап].[тип_параметра].[название_параметра]

Пример: rougher.input.feed_ag

Возможные значения для блока [этап]:
- rougher — флотация
- primary_cleaner — первичная очистка
- secondary_cleaner — вторичная очистка
- final — финальные характеристики

Возможные значения для блока [тип_параметра]:
- input — параметры сырья
- output — параметры продукта
- state — параметры, характеризующие текущее состояние этапа
- calculation — расчётные характеристики

### Произведем валидацию данных

**Расчёт эффективности**
![title](https://pictures.s3.yandex.net/resources/Recovery_1576238822.jpg)

где:
- C — доля золота в концентрате после флотации/очистки (`rougher.output.concentrate_au`);
- F — доля золота в сырье/концентрате до флотации/очистки (`rougher.input.feed_au`);
- T — доля золота в отвальных хвостах после флотации/очистки (`rougher.output.tail_au`).

```python
calculations = gold_recovery_train.copy()[['rougher.output.concentrate_au',
                                           'rougher.input.feed_au',
                                           'rougher.output.tail_au',
                                           'rougher.output.recovery']]
```

```python
C = calculations['rougher.output.concentrate_au']
F = calculations['rougher.input.feed_au']
T = calculations['rougher.output.tail_au']
```

```python
calculations['recovery_calculate'] = (C*(F-T))/(F*(C-T))*100
```

```python
calculations.head()
```

```python
calculations = calculations.dropna()
```

```python
print('Абсолютная средняя ошибка =', mean_absolute_error(calculations['rougher.output.recovery'],
                                                         calculations['recovery_calculate'])
     )
```

### Напишем функцию для вычисления метрики sMAPE и проверим ее

```python
def sMAPE(y_test, y_pred):
    '''Рассчитывает sMAPE флотации, где numerator - числитель, а denominator - знаменатель'''
    numerator_roughe = abs(y_test[y_test.columns[0]] - y_pred[:,0])
    denominator_roughe = (abs(y_test[y_test.columns[0]]) + abs(y_pred[:,0]))/2 
    score_roughe = (numerator_roughe / denominator_roughe).mean() *100
    '''Рассчитывает sMAPE черного концентрата, где numerator - числитель, а denominator - знаменатель'''
    numerator_final = abs(y_test[y_test.columns[1]] - y_pred[:,1])
    denominator_final = (abs(y_test[y_test.columns[1]]) + abs(y_pred[:,1]))/2 
    score_final = (numerator_final / denominator_final).mean() *100
    '''Итоговый sMAPE'''
    sMAPE_score = 0.25 * score_roughe + 0.75 * score_final
    return sMAPE_score
```
